```python
df1 = pd.read_excel("estate.xlsx",parse_dates=["transaction_date"],index_col = "No")
```

```python
df1.head()
```

```python
df1.drop(columns=['transaction_date'],inplace = True)
```

```python
X = df1.iloc[:, :-1].values
y = df1.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
model1 = tf.keras.models.Sequential([
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(512,activation=tf.nn.elu),
tf.keras.layers.Dropout(0.2),
#tf.keras.layers.Dense(512,activation=tf.nn.elu),
tf.keras.layers.Dense(512,activation=tf.nn.elu),
tf.keras.layers.Dropout(0.2),
tf.keras.layers.Dense(125,activation=tf.nn.elu),
tf.keras.layers.Dense(64,activation=tf.nn.elu),
#tf.keras.layers.Dense(15,activation = tf.nn.tanh),
tf.keras.layers.Dense(1)#,activation=tf.nn.sigmoid)
])
```

```python
epochs = 100
batch_size = 16
```

```python
optimiser = tf.keras.optimizers.Adam(lr = 0.01)
model1.compile (optimizer= optimiser, loss='mean_squared_error')#, metrics = ['accuracy'])
model1.fit(X_train,y_train,epochs = epochs,batch_size = batch_size)
```

```python
y_pred = model1.predict(X_test)
```
